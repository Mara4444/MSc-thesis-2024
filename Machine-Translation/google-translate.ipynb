### Create translated MGSM benchmark file for language list 

```python
import googletrans
from googletrans import Translator
import pandas as pd

translator = Translator()
```

```python
print(googletrans.LANGUAGES) # available languages
print(googletrans.LANGUAGES['af'])
```

```python
def translate_mgsm(langlist):
    """
    Translate the English msgsm benchmark to the languages in the input list
    and saving to tsv file.
    
    Parameters:
    langlist: list of the languages of interest, languages given in iso-2 code.
    
    Returns:
    saved mgsm_<lang>.tsv file
    """
    mgsm = pd.read_csv('./mgsm_original/mgsm_English.tsv',sep='\t', header=None) # the English version of MGSM benchmark
    
    questionlist_input = mgsm.loc[:,0] # question column
    answerlist_input = mgsm.loc[:,1] # answer column

    questionlist_output = []

    for lang in langlist:
        for question in questionlist_input:
            translation = translator.translate(question, src='en', dest=lang)
            questionlist_output.append(translation.text)

        mgsm_output = pd.DataFrame(data=[row.split(sep="',") for row in questionlist_output]) # converting the translated questionlist to a pandas df
        mgsm_output = pd.concat([mgsm_output, answerlist_input], axis=1) # adding the answer column

        mgsm_output.to_csv('./mgsm_translated/mgsm_' + googletrans.LANGUAGES[lang].title() + '.tsv', sep='\t', index=False, header=False) # or without answers? or as csv with ;?
```

```python
languages = ['af']

# approx 5 min per language
translate_mgsm(languages)
```
